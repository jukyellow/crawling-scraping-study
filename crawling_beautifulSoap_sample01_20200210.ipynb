```
# 1. hello crawling
import requests

URL = 'https://unipass.customs.go.kr/clip/index.do'
VERIFY = False

## HTTP GET Request
req = requests.get(URL, verify=VERIFY)

## HTML 소스 가져오기
html = req.text
print('html:', html)

## HTTP Header 가져오기
header = req.headers
print('header:', header)

## HTTP Status 가져오기 (200: 정상)
status = req.status_code
print('status:', status)

## HTTP가 정상적으로 되었는지 (True/False)
is_ok = req.ok
print('is_ok:', is_ok)
```

```
#참고2:
# 2. get contents
import requests
from bs4 import BeautifulSoup

def getReq():
    URL = 'https://unipass.customs.go.kr/clip/index.do'
    VERIFY = False
    req = requests.get(URL, verify=VERIFY)
    return req

def getSoup(req):
    soup = BeautifulSoup(req.content, 'html.parser')
    return soup

def checkConn(req):
    ## HTTP Header 가져오기
    header = req.headers
    #print('header:', header)

    ## HTTP Status 가져오기 (200: 정상)
    status = req.status_code
    print('status:', status)

    ## HTTP가 정상적으로 되었는지 (True/False)
    is_ok = req.ok
    #print('is_ok:', is_ok)
```

```
#extract sample
req = getReq()
checkConn(req)
soup = getSoup(req)

# 태그 검색
title = soup.find('title') 
print('title:', title)
print ('title.title():', title.get_text())

# select 함수는 리스트 형태로 전체 반환
select_tag = soup.select('a')
print (select_tag)

## my_titles는 list 객체
for title in select_tag:    
    print(title.text) ## Tag안의 텍스트  
```
